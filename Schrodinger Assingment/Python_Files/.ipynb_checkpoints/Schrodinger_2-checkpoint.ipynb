## Note about units
We will be using atomic units throughout.

```python
# atomic units
hbar = 1.0
m = 1.0
#set precision of numerical approximation
steps = 2000
```

## Particle in an infinite potential well

```python
########
# PARTICLE IN AN INFINITE POTENTIAL WELL
########
Case = 1
########
# INPUT
########
# set well width and number of wave function desired
W = 1.0
n = 7
########
# CODE
########
# divide by two so a well from -W to W is of input width
W = W / 2.0
# create x-vector from -W to W
xvec = np.linspace(-W,W,steps,dtype=np.float_)
h = xvec[1]-xvec[0]
# create Laplacian via 3 point finite-difference method
Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)+np.diag(np.ones(steps-1),-1))/(float)(h**2)
# create Hamiltonian
Hamiltonian=((-0.5*(hbar**2)/m))*Laplacian
# diagonalize the Hamiltonian yielding the wavefunctions and Energies
E , V = diagonalize_hamiltonian(Hamiltonian)
# print output
output(Case,['Well Width','Number of wavefunctions to plot'],[W,n],E,n)
# create plot
infinite_well_plot(E,V,xvec,W,steps,n,Case)
```

## Particle in a finite well

```python
########
# PARTICLE IN A FINITE WELL OF WIDTH (W) AND DEPTH (D)
########
Case = 2
########
# INPUT
########
# set well depth and width
A  = 1.0
D = -100.0
########
# CODE
########
# divide by two so a well from -W to W is of input width
W = A / 2.0
# create x-vector from -W to W
xvec = np.linspace(-A,A,steps,dtype=np.float_)
h = xvec[1]-xvec[0]
# create the potential from step function
U=D*(step_func(xvec+W)-step_func(xvec-W))
# create Laplacian via 3-point finite-difference method
Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)+np.diag(np.ones(steps-1),-1))/(float)(h**2)
# create the Hamiltonian
Hamiltonian = np.zeros((steps,steps))
[i,j]=np.indices(Hamiltonian.shape)
Hamiltonian[i==j]=U
Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian
# diagonalize the Hamiltonian yielding the wavefunctions and Energies
E , V = diagonalize_hamiltonian(Hamiltonian)
# Determine number of energy levels to plot (n)
n = 0
while E[n] < 0:
    n+=1
# print output
output(Case,['Well Width','Well Depth'],[W,D],E,n)
finite_well_plot(E,V,xvec,steps,n,Case,U)
```

## Particle in a double finite well (equal depth)

```python
########
# PARTICLE IN A DOUBLE FINITE WELL OF WIDTH (W), DISTANCE (B) APART, AND WELLS OF EQUAL DEPTH (D)
########
Case = 3
########
# INPUT
########
#set depths and widths of wells and the well separation
W = 1.0
D = -60.0
B = 0.6
########
# CODE
########
A=2.0*((2.0*W) + B)   # Set length variable for xvec
B = B / 2.0
# create x-vector from -W to W
xvec = np.linspace(-A,A,steps,dtype=np.float_)
h = xvec[1]-xvec[0]
U=D*(step_func(xvec+W+B)-step_func(xvec+B)+step_func(xvec-B)-step_func(xvec-W-B))   # Create the potential from stepfunc
# create Laplacian via 3-point finite-difference method
Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)+np.diag(np.ones(steps-1),-1))/(float)(h**2)
Hamiltonian = np.zeros((steps,steps))
[i,j]=np.indices(Hamiltonian.shape)
Hamiltonian[i==j]=U
Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian
# diagonalize the Hamiltonian yielding the wavefunctions and Energies
E , V = diagonalize_hamiltonian(Hamiltonian)
# Determine number of energy levels to plot (n)
n = 0
while E[n] < 0:
    n+=1
# print output
output(Case,['Well Widths','Well Depths','Well Separation'],[W,D,B],E,n)
finite_well_plot(E,V,xvec,steps,n,Case,U)
```

## Particle in a double finite well (unequal depth)

```python
########
# PARTICLE IN A DOUBLE FINITE WELL OF WIDTH (W) AND DISTANCE (B) APART WITH WELLS OF DIFFERENT DEPTHS (D1 and D2)
########
Case=4
########
# INPUT
########
#set depths and widths of wells and the well separation
W1 = 1.0
W2 = 0.5
D1 = -60.0
D2 = -70.0
B = 0.2
########
# CODE
########
A=2.0*((W1+W2) + B)   # Set length variable for xvec
B = B / 2.0
# create x-vector from -W to W
xvec = np.linspace(-A,A,steps,dtype=np.float_)
h = xvec[1]-xvec[0]
U = D1*(step_func(xvec+W1+B)-step_func(xvec+B))+D2*(step_func(xvec-B)-step_func(xvec-W2-B))   # Create the potential from stepfunc
Laplacian = (-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)+np.diag(np.ones(steps-1),-1))/(float)(h**2)
Hamiltonian=np.zeros((steps,steps))
[i,j]=np.indices(Hamiltonian.shape)
Hamiltonian[i==j]=U
Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian   # Calculate the Hamiltonian
# diagonalize the Hamiltonian yielding the wavefunctions and Energies
E , V = diagonalize_hamiltonian(Hamiltonian)
# Determine number of energy levels to plot (n)
n = 0
while E[n] < 0:
    n+=1
# print output
output(Case,['Well 1 Width','Well 2 Width','Well 1 Depth','Well 2 Depth','Well Separation'],[W1,W2,D1,D2,B],E,n)
finite_well_plot(E,V,xvec,steps,n,Case,U)
```

## Particle in Harmonic Well

```python
########
# PARTICLE IN A HARMONIC WELL OF DEPTH (D) WITH A FORCE CONSTANT (omega)
########
Case = 5
########
# INPUT
########
# set force constant and depth of harmonic well
omega = 1.5
D = -7
########
# CODE
########
# divide by two so a well from -W to W is of input width
W = np.sqrt(np.abs(2.0 * D)/(omega**2))
A = W * 2.0
# create x-vector from -W to W
xvec = np.linspace(-A,A,steps,dtype=np.float_)
h = xvec[1]-xvec[0]
# create the potential from step function
U=harmonic_potential(xvec,omega,D)
# create Laplacian via 3-point finite-difference method
Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)+np.diag(np.ones(steps-1),-1))/(float)(h**2)
# create the Hamiltonian
Hamiltonian = np.zeros((steps,steps))
[i,j]=np.indices(Hamiltonian.shape)
Hamiltonian[i==j]=U
Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian
# diagonalize the Hamiltonian yielding the wavefunctions and Energies
E , V = diagonalize_hamiltonian(Hamiltonian)
# Determine theoretical number of energy levels (n)
n = 0
while E[n] < 0:
    n+=1
# print output
output(Case,['Force Constant','Depth'],[omega,D],E,n)
finite_well_plot(E,V,xvec,steps,n,Case,U)
```

```python
########
# PARTICLE IN A MORSE WELL OF DEPTH (D) WITH A FORCE CONSTANT (omega)
########
Case = 6
########
# INPUT
########
# set force constant and depth of morse well
omega = 0.1
D = 5.0
########
# CODE
########
# create the potential
xvec,h,U = morse_potential(omega,D,steps)
h = xvec[1]-xvec[0]
# create Laplacian via 3-point finite-difference method
Laplacian=(-2.0*np.diag(np.ones(steps))+np.diag(np.ones(steps-1),1)+np.diag(np.ones(steps-1),-1))/(float)(h**2)
# create the Hamiltonian
Hamiltonian = np.zeros((steps,steps))
[i,j]=np.indices(Hamiltonian.shape)
Hamiltonian[i==j]=U
Hamiltonian+=(-0.5)*((hbar**2)/m)*Laplacian
# diagonalize the Hamiltonian yielding the wavefunctions and Energies
E , V = diagonalize_hamiltonian(Hamiltonian)
# Determine theoretical number of energy levels (n)
n = 0
while E[n] < 0:
    n+=1
# print output
output(Case,['Force Constant','Depth'],[omega,D],E,n)
finite_well_plot(E,V,xvec,steps,n,Case,U)
```
